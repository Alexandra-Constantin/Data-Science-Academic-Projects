## Generate random vectors and calculate correlations (6p)

In general, two vectors (of the same length and of normally distributed random numbers) should have their Pearson correlation close to zero.  
Let's test that.

Many times, generate two random vectors of the same size and calculate their correlation. Find the mean and the standard deviation of the correlations.  
Repeat the calculation for different sizes of the vectors.  
Finally, present the results in a table as shown below (with a header, with numbers rounded, columns separated with `\t` tabulator).
Here is a fragment of a result (to demonstrate the format, the numbers are random):

```text
vecSize	meanCor	sdCor
...
20	0.0075	0.22274
50	-0.0218	0.13385
100	0.0062	0.10085
...
```

Requirements:

- Study vector sizes `vecSizes=[10, 20, 50, 100, 200, 500, 1000, 2000, 5000]`.
- For each `vecSize` generate `repeatNum=100` pairs of random vectors, and calculate their Pearson correlation.
- Use only the functions discussed in the Python base lectures 01-05 (libraries `random` and `statistics` are ok, but no `numpy` nor `pandas`).
- Use simple formatted printing (no `pandas` or `tabulate`).

````python
# ----- SOLUTION START -----

import random
import statistics

repeatNum = 100
vecSizes = [10, 20, 50, 100, 200, 500, 1000, 2000, 5000]

# define function
def pearson_corr(x, y):
    mean_x = statistics.mean(x)
    mean_y = statistics.mean(y)
    
    numerator = sum( ((x[i]) - mean_x ) * (y[i] - mean_y) for i in range(len(x)) )
    denom_x = sum( (x[i] - mean_x)**2 for i in range(len(x)) ) **0.5
    denom_y = sum( (y[i] - mean_y)**2 for i in range(len(y)) ) **0.5
    
    if denom_x == 0 or denom_y == 0:
        return 0
    else:
        return numerator / (denom_x * denom_y)

results = []

# generate random vectors and compute correlations
for size in vecSizes:
    correlations = []
    
    for i in range(repeatNum):
        x = []
        y = []
        
        for i in range(size):
            x.append(random.gauss(0, 1))
            y.append(random.gauss(0, 1))
        
        corr = pearson_corr(x, y)  # compute correlations
        correlations.append(corr)  # store correlations
    
    # compute mean and SD of correlations
    mean_corr = statistics.mean(correlations)
    sd_corr = statistics.stdev(correlations)
    results.append((size, mean_corr, sd_corr))
    
#print results in table
print("vecSize\tmeanCor\tsdCor")
for size, mean, sd in results:
    print(f"{size}\t{mean_corr:.4f}\t{sd_corr:.5f}")


# ----- SOLUTION END -----
````

````python
# ----- SOLUTION START -----

import math

class Vec:
    def __init__(self, x = 0, y = 0):
        """ Initialize vector with x, y coordinates """
        self._x = x
        self._y = y

    def x(self):
        """ Returns x-coordinate of vector """
        return self._x

    def y(self):
        """ Returns y-coordinate of vector """
        return self._y

    def len(self):
        """ Returns the length of the vector based on Euclidean distance """
        return math.sqrt(self._x **2 + self._y **2)

    def deg(self):
        """ Returns the angle between the positive x-axis and the direction of the vector """
        return math.degrees(math.atan2(self._y, self._x))

    def add(self, v):
        """ Adds another vector `v` to current vector `self` """
        self._x += v.x()
        self._y += v.y()
        return self
    
    def mul(self, n):
        """ Multiplies the vector a scalar """
        self._x *= n
        self._y *= n
        return self
    
    def rotate(self, deg):
        """ Rotates the vector by given nr. of degrees """ 
        rad = deg * (math.pi / 180)
        x_rotated = self._x * math.cos(rad) - self._y * math.sin(rad)
        y_rotated = self._x * math.sin(rad) + self._y * math.cos(rad)
        return self

    def __str__(self):
        """ Returns a string representation of the vector """
        return f" Vec(x = {self._x}, y = {self._y} ; len = {self.len()}, deg = {self.deg()}) "


# ----- SOLUTION END -----
````

## RPN (reverse polish notation) calculator (exceptions, flow control, list, stack) (7p)

[Reverse polish notation](https://en.wikipedia.org/wiki/Reverse_Polish_notation) allows to write mathematical expressions without need of `(` and `)`. Consider the examples:

| RPN notation tokens | "Normal" notation | Result |
| ----- | ----- | ----- |
| `1`     | `1` | 1 |
| `1` `2.5` `+` | `1 + 2.5` | 3.5 |
| `1` `2` `3` `*` `+` | `1 + 2 * 3` | 7 |
| `1` `2` `3` `*` `+` | `1 + (2*3)` | 7 |
| `1` `2` `+` `3` `*` | `(1+2) * 3` | 9 |

Before reading further, understand the concept of a [stack](https://en.wikipedia.org/wiki/Stack_(abstract_data_type)).  
In RPN each subsequent argument (`token`) is checked:
- when it is a number: 
    - the number is put on the stack (`push`)
- when it is an operator (`+` addition, `-` subtraction, `*` multiplication, `/` division):
    - two recent numbers are removed from the stack (2*`pop`)
    - the calculation specified by the operator is performed
    - the result is pushed to the stack (`push`)

Write a function `rpn(tokens)` which takes a list of tokens (e.g. `[ 1, 2, "+" ]`) and returns a number - the result of the calculation.  
There are several errors possible - the function should raise an exception with a message clearly describing the problem (see errors below).

Some example calls of the function and their expected effects (other examples will be used for grading):
```python
print( rpn( [ 1 ] ) )                             # 1
print( rpn( [ 1, 2.5, "+" ] ) )                   # 3.5 i.e. 1+2.5
print( rpn( [ -1, 2, 3, "+", "*" ] ) )            # -5 i.e. -1*(2+3)
print( rpn( [ 5, 7, "+", 2, 1, "+", "/" ] ) )     # 4 i.e. (5+7)/(2+1)
# print( rpn( [ 1, "+" ] ) )                      # RuntimeError: Not enough arguments for + operator.
# print( rpn( [ 1, 2 ] ) )                        # RuntimeError: Not enough operators; too many elements on remained on stack.
# print( rpn( [ "a" ] ) )                         # ValueError: could not convert string to float: 'a'
```

*Hint:* Use Python `list` as a stack (methods: `push` is `append(...)`, `pop` is  `pop(-1)`; function `len(...)`).

````python
# ----- SOLUTION START -----
def rpn(tokens):
    stack = []
        
    for token in tokens:
        if isinstance(token, int) or isinstance(token, float):
            stack.append(token)
                
        elif token in ["+", "-", "*", "/"]:
            if len(stack) < 2:
                raise RuntimeError (f"Not enough arguments for {token} operator.") 
            
            b = stack.pop()
            a = stack.pop()
            
            if token == "+":
                stack.append(a+b)
            
            if token == "-":
                stack.append(a-b)
            
            if token == "*":
                stack.append(a*b)
            
            if token == "/":
                if b == 0:
                    raise ZeroDivisionError ("Division by zero not allowed.")
                
                stack.append(a/b)
        
        else:
            raise ValueError (f"Could not convert string to float {token}.")

    if len(stack) > 1:
        raise RuntimeError (f"Not enough arguments for {token} operator.")

    return stack[0]


# ----- SOLUTION END -----
````

````python
print( rpn( [ 1 ] ) )                             # 1
print( rpn( [ 1, 2.5, "+" ] ) )                   # 3.5 i.e. 1+2.5
print( rpn( [ -1, 2, 3, "+", "*" ] ) )            # -5 i.e. -1*(2+3)
print( rpn( [ 5, 7, "+", 2, 1, "+", "/" ] ) )     # 4 i.e. (5+7)/(2+1)
print( rpn( [ 1, "+" ] ) )                      # RuntimeError: Not enough arguments for + operator.
print( rpn( [ 1, 2 ] ) )                        # RuntimeError: Not enough operators; too many elements on remained on stack.
print( rpn( [ "a" ] ) )                         # ValueError: could not convert string to float: 'a'
````

````python
# ----- SOLUTION START -----

import sys

sys.argv

def rpn(tokens):
    stack = []
        
    for token in tokens:
        
        try:
            stack.append(float(token))
        except ValueError:
            pass
                
            if token in ["+", "-", "*", "/"]:
                if len(stack) < 2:
                    raise RuntimeError (f"Not enough arguments for {token} operator.") 
            
                b = stack.pop()
                a = stack.pop()
            
                if token == "+":
                    stack.append(a+b)
            
                if token == "-":
                    stack.append(a-b)
            
                if token == "*":
                    stack.append(a*b)
            
                if token == "/":
                    if b == 0:
                        raise ZeroDivisionError ("Division by zero not allowed.")
                
                    stack.append(a/b)
        
            else:
                raise ValueError (f"Could not convert string to float {token}.")

        if len(stack) > 1:
            raise RuntimeError (f"Not enough arguments for {token} operator.")

        return stack[0]

if __name__ == "__main__":
    tokens = sys.argv[1:]
    print(rpn(tokens))


# ----- SOLUTION END -----
````
